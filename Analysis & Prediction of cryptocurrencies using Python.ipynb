```python
import pandas_datareader.data as reader
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
end = dt.datetime.now()
start = dt.date(end.year - 4,end.month,end.day)

cryptolist = ['BTC-USD', 'ETH-USD', 'XRP-USD', 'LTC-USD']
```

```python
df = reader.get_data_yahoo(cryptolist,start,end)['Adj Close']
```

```python
df
```

```python
df.plot()
plt.show()
```

```python
 daily_returns = df.pct_change().dropna(axis=0)
```

```python
daily_returns
```

```python
daily_returns.plot()
plt.show()
```

```python
daily_cum_returns = (daily_returns + 1).cumprod() -1
```

```python
daily_cum_returns
```

```python
daily_cum_returns.plot()
plt.show()
```

```python
colors = ['r', 'k', 'b', 'g']
daily_cum_returns.plot(color=colors, figsize=(15,6))
plt.title('Cumlative returns of the largest cryptocurrencies')
plt.show()
                       
```

```python
fig, axs = plt.subplots(2, 2,figsize=(15,8),gridspec_kw={'hspace': 0.2, 'wspace': 0.1})

axs[0, 0].plot(df['BTC-USD'], c='r')
axs[0, 0].set_title('BTC')
axs[0, 1].plot(df['ETH-USD'], c='k')
axs[0, 1].set_title('ETH')
axs[1, 0].plot(df['XRP-USD'], c='b')
axs[1, 0].set_title('XRP')
axs[1, 1].plot(df['LTC-USD'], c='g')
axs[1, 1].set_title('LTC')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2,figsize=(15,8),gridspec_kw={'hspace': 0.2, 'wspace': 0.1})

axs[0, 0].hist(daily_returns['BTC-USD'], bins=100,color='r', range=(-0.2, 0.2))
axs[0, 0].set_title('BTC')
axs[0, 1].hist(daily_returns['ETH-USD'], bins=100,color='k', range=(-0.2, 0.2))
axs[0, 1].set_title('ETH')
axs[1, 0].hist(daily_returns['XRP-USD'], bins=100,color='b', range=(-0.2, 0.2))
axs[1, 0].set_title('XRP')
axs[1, 1].hist(daily_returns['LTC-USD'], bins=100,color='g', range=(-0.2, 0.2))
axs[1, 1].set_title('LTC')
plt.show()
```

```python
from fbprophet import Prophet
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
end1 = dt.date(end.year - 1,end.month,end.day) #1 year back
end2 = dt.date(end.year,end.month - 6,end.day) #6 month back
start1 = dt.date(end1.year - 3,end.month,end.day)
```

```python
df1 = reader.get_data_yahoo(cryptolist,start1,end1)['Adj Close']
df2 = reader.get_data_yahoo(cryptolist,start1,end2)['Adj Close']
```

```python
modelfb1 = Prophet()
df1 = df1.reset_index()
df1[['ds', 'y']] = df1[['Date', 'BTC-USD']]
modelfb1.fit(df1)
```

```python
modelfb2 = Prophet()
df2 = df1.reset_index()
df2[['ds', 'y']] = df1[['Date', 'BTC-USD']]
modelfb2.fit(df2)
```

```python
future1 = modelfb1.make_future_dataframe(periods=365)
```

```python
future2 = modelfb1.make_future_dataframe(periods=180)
```

```python
forecast1 = modelfb1.predict(future1)
forecast2 = modelfb2.predict(future2)
```

```python
modelfb1.plot(forecast1)
plt.title('Backtest of BTC 1 Year Prediction')
plt.show()
```

```python
modelfb2.plot(forecast2)
plt.title('Backtest of BTC 6 month Prediction')
plt.show()
```
